```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed data
data = pd.read_csv("preprocessed_heart.csv")

# Step 1: Visualize the distribution of target variable
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=data, palette='coolwarm')
plt.title('Distribution of Target Variable')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Step 2: Visualize the correlation matrix
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Step 3: Pairplot for key features
key_features = ['age', 'chol', 'thalach', 'trestbps', 'target']  # Adjust features as needed
sns.pairplot(data[key_features], hue='target', palette='coolwarm')
plt.title('Pairplot of Key Features')
plt.show()

# Step 4: Boxplot for age vs. target
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='age', data=data, palette='coolwarm')
plt.title('Boxplot of Age vs. Target')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()

# Step 5: Histogram of key numerical features
numerical_features = ['age', 'chol', 'thalach', 'trestbps']
data[numerical_features].hist(bins=15, figsize=(12, 10), color='steelblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```
